```python

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait

from time import sleep
```

```python

service = Service(
    ChromeDriverManager().install()
    )
driver = webdriver.Chrome(service=service)

driver.get(
    "https://app.lenarge.com.br/"    )
driver.maximize_window()
```

```python
buttons_login = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, 'input'))
)

if len(buttons_login) >= 2:
    buttons_login[0].send_keys("kayky.soares@lenarge.com.br")
    buttons_login[1].send_keys("kyra0580")

try:
    driver.find_element(By.ID, "login-submit").click()
except Exception as e:
    print (e)
```

```python
driver.find_element(
    By.XPATH,
    '//a[span[contains(text(), "Operação")]]'
        ).click() 
driver.find_element(
    By.XPATH,
    '//div[@role="rowgroup"]/div[@role="row"]//div[contains(@class, "control")]'
        ).click() 
driver.find_element(
    By.XPATH,
    '//button[@class="button is-circle is-elevated is-primary m-l-10"]'
        ).click()
driver.find_element(
    By.XPATH,
    '(//div[contains(@class, "h-select") and contains(@class, "has-loader")]//div[contains(@class, "select-box")][span[normalize-space(text())="Placas e Motorista"]])'
).click()
```

```python
wait = WebDriverWait(driver, 15)

element = wait.until(EC.presence_of_element_located((
    By.XPATH,
    '//div[@class="option-row"]/div[@class="option-meta"][span[normalize-space(text())="Completo"]]'
)))

# Faz scroll suave até o elemento (garante visibilidade)
driver.execute_script(
    "arguments[0].scrollIntoView({block: 'center'});", element
    )

sleep(1)

wait.until(EC.element_to_be_clickable((
    By.XPATH,
    '//div[@class="option-row"]/div[@class="option-meta"][span[normalize-space(text())="Completo"]]'
)))

driver.execute_script("arguments[0].click();", element)
```
